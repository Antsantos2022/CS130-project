```python
import pandas as pd
import matplotlib.pyplot as plt 

url = "https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2021/2021-03-02/youtube.csv"
df = pd.read_csv(url) 

print(df.head())
print(df.info())

# Top brands in the ads
top_brands = df['brand'].value_counts().head(10)
plt.figure()
top_brands.plot(kind='bar')
plt.title('Top 10 Brands')
plt.xlabel('Brand')
plt.ylabel('Amount')
plt.tight_layout()
plt.savefig('viz1_top_brands.png')
plt.show() 


# Funny Superbowl Ads overtime 
funny_by_year = df.groupby('year')['funny'].mean() * 100
plt.figure()
plt.plot(funny_by_year.index, funny_by_year.values)
plt.title('Funny Ads Over Time')
plt.xlabel('Year')
plt.ylabel('Percent') 
plt.tight_layout()
plt.savefig('viz2_funny_time.png')
plt.show()

# Funny Ads vs Non Funny Ads 
funny_views = df[df['funny'] == True]['view_count'].mean()
not_funny_views = df[df['funny'] == False]['view_count'].mean()

plt.figure()
plt.bar(['Funny', 'Not Funny'], [funny_views, not_funny_views])
plt.title('Average Views: Funny vs Not')
plt.ylabel('Average Views') 
plt.tight_layout()
plt.savefig('viz3_funny_views.png')
plt.show()


# Views vs Likes 
plt.figure()
plt.scatter(df['view_count'], df['like_count'])
plt.title('Views vs Likes')
plt.xlabel('Views')
plt.ylabel('Likes')
plt.tight_layout()
plt.savefig('viz4_views_likes.png')
plt.show()



```
